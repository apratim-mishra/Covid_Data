```python
time_male = time_gender[time_gender['sex']=='male']
```

```python
time_female = time_gender[time_gender['sex']=='female']
```

```python
time_male
```

```python
time_male = time_male.rename(columns={'confirmed':'confirmed_male', 'deceased':'deceased_male'})
```

```python
time_male.reset_index(inplace=True, drop=True)
```

```python
time_female = time_female.rename(columns={'confirmed':'confirmed_female', 'deceased':'deceased_female'})
```

```python
time_female.reset_index(inplace=True, drop=True)
```

```python
time_female
```

```python
pd.concat([time_male, time_female], axis=1)
```

```python
time_newgender = pd.concat([time_male, time_female], axis=1)
```

```python
time_newgender.drop(columns='time', inplace=True)
```

```python
time_newgender.drop(columns='sex', inplace=True)
```

```python
time_newgender
```

```python
time_newgender.tail(4)
```

```python
time_newgender = time_newgender.loc[:, ~time_newgender.columns.duplicated()]
```

```python
time_newgender
```

```python
time_newgender['Events'] = np.nan
```

```python
time_newgender.loc[time_newgender['date']=='2020-03-02', 'Events'] ='Lockdown'
```

```python
time_newgender.loc[time_newgender['date']=='2020-04-17', 'Events'] ='Restrictions relaxed'
```

```python
time_newgender.loc[time_newgender['date']=='2020-05-20', 'Events'] ='Initial Opening'
```

```python
time_newgender.loc[time_newgender['date']=='2020-05-28', 'Events'] ='Restrictions reinitiated'
```

```python
time_newgender.to_csv('time_gender.csv')
```
