```python
dates=global_conf.columns[4:]
```

```python
confirmed = global_conf.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Confirmed')
```

```python
deaths = global_deaths.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Deaths')
```

```python
recovered = global_recv.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], 
                            value_vars=dates, var_name='Date', value_name='Recovered')
```

```python
confirmed.shape, deaths.shape, recovered.shape
```

```python
combined = pd.merge(left=combined_df, right=recovered, how='left', on=['Province/State', 'Country/Region', 'Date', 'Lat', 'Long'])
```

```python
combined['Recovered'].isna().any()
```

```python
combined['Recovered'] = combined['Recovered'].fillna(0)
```

```python
combined['Active Cases'] = combined['Confirmed'] - combined['Deaths'] - combined['Recovered']
```

```python
combined['Active Cases'] = combined['Active Cases'].apply(lambda x: 0.0 if x<0 else x)
```
